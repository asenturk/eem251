```python
import matplotlib.pyplot as plt
```

```python
a=list(range(1,10))
b=[i**2 for i in a]
print(a)
print(b)
```

```python
plt.plot(a,b)
plt.show()
```

```python
import numpy as np
a=np.arange(1,10,0.1)
b=a**2
plt.plot(a,b)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
sinx=np.sin(x)
plt.plot(x,sinx)
#plt.show()
cosx=np.cos(x)
plt.plot(x,cosx)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
sinx=np.sin(x)

plt.title("sin ve cos sinyalleri")

plt.plot(x,sinx)
#plt.show()
cosx=np.cos(x)
plt.plot(x,cosx)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
sinx=np.sin(x)

plt.title("sin ve cos sinyalleri")
plt.xlabel("X")
plt.ylabel("sin(x), cos(x)")

plt.plot(x,sinx)
#plt.show()
cosx=np.cos(x)
plt.plot(x,cosx)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
sinx=np.sin(x)

plt.title("sin ve cos sinyalleri")
plt.xlabel("X")
plt.ylabel("sin(x), cos(x)")

plt.plot(x,sinx, label="sin", color="red")
#plt.show()
cosx=np.cos(x)
plt.plot(x,cosx, label="cos", color="blue")
plt.legend()
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
sinx=np.sin(x)

plt.title("sin ve cos sinyalleri")
plt.xlabel("X")
plt.ylabel("sin(x), cos(x)")

plt.plot(x,sinx, "--r", label="sin")
#plt.show()
cosx=np.cos(x)
plt.plot(x,cosx, "-.g", label="cos")
plt.legend()
plt.show()
```
